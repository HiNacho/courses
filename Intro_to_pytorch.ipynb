## What are Tensors?

In PyTorch, we use something called **tensors**. You can think of tensors as special containers that hold numbers.

*   A tensor can be a single number, like holding the number of apples you have (a 0-dimensional tensor or scalar).
*   It can be a list of numbers, like the scores you got on a test (a 1-dimensional tensor or vector).
*   It can be a table of numbers with rows and columns, like a spreadsheet (a 2-dimensional tensor or matrix).
*   It can even be more complex, like a box of tables (higher-dimensional tensors)!

Tensors are like the building blocks we use in PyTorch to store and work with data.

## Welcome to PyTorch Basics

PyTorch is a powerful tool used for working with numbers and data. Think of it like a special way to organize and do math with lots of numbers at once!

We use PyTorch to help computers learn and understand things, like recognizing patterns in data. It's used for various tasks in areas like artificial intelligence and machine learning.

In this notebook, we're going to learn some fundamental concepts of PyTorch by looking at some code examples.

```python
import torch

# Create a tensor
original_tensor = torch.arange(12).reshape(3, 4)
print("Original tensor:")
print(original_tensor)

# Use view to create a new tensor
view_tensor = original_tensor.view(4, 3)
print("\nTensor after using view:")
print(view_tensor)

# Modify the view tensor
view_tensor[0, 0] = 99
print("\nView tensor after modification:")
print(view_tensor)
print("\nOriginal tensor after modifying view tensor:")
print(original_tensor) # Note that the original tensor is also changed
```

```python
import torch

# Create a tensor (re-create to demonstrate reshape behavior independently)
original_tensor_2 = torch.arange(12).reshape(3, 4)
print("Original tensor 2:")
print(original_tensor_2)

# Use reshape to create a new tensor
reshape_tensor = original_tensor_2.reshape(4, 3)
print("\nTensor after using reshape:")
print(reshape_tensor)

# Modify the reshape tensor
reshape_tensor[0, 0] = 99
print("\nReshape tensor after modification:")
print(reshape_tensor)
print("\nOriginal tensor 2 after modifying reshape tensor:")
print(original_tensor_2) # Note that the original tensor is also changed in this case, as reshape returned a view
```
