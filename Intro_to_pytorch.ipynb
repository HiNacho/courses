## What are Tensors?

In PyTorch, we use something called **tensors**. You can think of tensors as special containers that hold numbers.

*   A tensor can be a single number, like holding the number of apples you have (a 0-dimensional tensor or scalar).
*   It can be a list of numbers, like the scores you got on a test (a 1-dimensional tensor or vector).
*   It can be a table of numbers with rows and columns, like a spreadsheet (a 2-dimensional tensor or matrix).
*   It can even be more complex, like a box of tables (higher-dimensional tensors)!

Tensors are like the building blocks we use in PyTorch to store and work with data.

## Welcome to PyTorch Basics

We use PyTorch to help computers learn and understand things, like recognizing patterns in data. It's used for various tasks in areas like artificial intelligence and machine learning.

### Building a Simple Neural Network Model

This code defines a simple neural network model using PyTorch's `nn.Module`. Think of a neural network like a series of interconnected layers that process information to make predictions or decisions.

Here's how this model works:

*   **`__init__`**: This is like the setup part of our model. We define the layers our network will have.
*   **`forward`**: This is where the data actually flows through the network. It defines the steps the data takes from the input layer through the hidden layers to the output layer.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Model(nn.Module):
  def __init__(self, in_features = 4, h1 = 8, h2 = 9, out_features =3):
    super().__init__() # Call the initializer of the parent class (nn.Module) to set up the basic structure of the module.
    self.fc1 = nn.Linear(in_features, h1) # Define the first fully connected layer. It takes 'in_features' as input and outputs 'h1' features.
    self.fc2 = nn.Linear(h1,h2) # Define the second fully connected layer. It takes 'h1' as input and outputs 'h2' features.
    self.out = nn.Linear(h2,out_features) # Define the output layer. It takes 'h2' as input and outputs 'out_features'.

  def forward(self,x):
    x = F.relu(self.fc1(x)) # Pass the input 'x' through the first layer and apply the Rectified Linear Unit (ReLU) activation function.
    x = F.relu(self.fc2(x)) # Pass the output of the first layer through the second layer and apply the ReLU activation function.
    x = self.out(x) # Pass the output of the second layer through the output layer.

    return x # Return the final output of the network.
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Model(nn.Module):
  def __init__(self, in_features = 4, h1 = 8, h2 = 9, out_features =3):
    super().__init__() # Call the initializer of the parent class (nn.Module) to set up the basic structure of the module.
    self.fc1 = nn.Linear(in_features, h1) # Define the first fully connected layer. It takes 'in_features' as input and outputs 'h1' features.
    self.fc2 = nn.Linear(h1,h2) # Define the second fully connected layer. It takes 'h1' as input and outputs 'h2' features.
    self.out = nn.Linear(h2,out_features) # Define the output layer. It takes 'h2' as input and outputs 'out_features'.

  def forward(self,x):
    x = F.relu(self.fc1(x)) # Pass the input 'x' through the first layer and apply the Rectified Linear Unit (ReLU) activation function.
    x = F.relu(self.fc2(x)) # Pass the output of the first layer through the second layer and apply the ReLU activation function.
    x = self.out(x) # Pass the output of the second layer through the output layer.

    return x # Return the final output of the network.
```

```python
torch.manual_seed(41) # Set the random seed for reproducibility

model = Model() # Create an instance of the Model class
```

```python
criterion = nn.CrossEntropyLoss() # Define the loss function for the model (Cross-Entropy Loss is suitable for classification tasks)

# Choose Adam Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.01) # Define the optimizer for updating the model's weights using the Adam algorithm with a learning rate of 0.01
```

```python
model.parameters # Access the parameters (weights and biases) of the model
```

```python
epochs = 100
losses = []

for i in range(epochs):
  y_pred = model.forward(X_train_tensor) # Perform a forward pass to get predictions
  loss = criterion(y_pred, y_train_tensor) # Calculate the loss between predictions and actual labels
  losses.append(loss.detach().numpy()) # Store the loss value for plotting

  if i % 10 == 0:
    print(f'Epoch: {i} Loss: {loss}') # Print the loss every 10 epochs

  optimizer.zero_grad() # Clear previous gradients
  loss.backward() # Perform backpropagation to calculate gradients
  optimizer.step() # Update the model's weights
```

```python
plt.plot(range(epochs), losses) # Plot the training loss over the epochs
plt.ylabel('Loss') # Label the y-axis as 'Loss'
plt.xlabel('epoch') # Label the x-axis as 'epoch'
```

```python
%%markdown
## Building a Simple Neural Network Model

Imagine you have a special machine that helps you sort things, like different colored balls. This machine has different steps inside to help it figure out where each ball should go.

Our **Neural Network Model** is kind of like that sorting machine! It's a set of steps that takes in numbers (our data) and processes them to give us an answer or a prediction.

Think of the `Model` class as the blueprint for our machine. It tells us what steps are inside and how they are connected.

### Setting Up the Machine (`__init__`)

The `__init__` part is like setting up our sorting machine. We define the different parts, or **layers**, it will have.

*   `super().__init__()`: This is like making sure our machine has a solid base to build on. `nn.Module` is that base provided by PyTorch, making it easier to build these machines.

*   `self.fc1 = nn.Linear(in_features, h1)`: This is our first sorting step. It takes the numbers coming *in* (`in_features` slots for numbers) and processes them to produce a new set of numbers (`h1` slots for numbers). It's like the first part of the machine that looks at the balls and starts to group them based on some initial rule.

*   `self.fc2 = nn.Linear(h1, h2)`: This is the second sorting step. It takes the numbers from the first step (`h1` slots) and processes them further to produce another set of numbers (`h2` slots). It's like the next part of the machine that refines the sorting based on another rule.

*   `self.out = nn.Linear(h2, out_features)`: This is the final sorting step, the output layer. It takes the numbers from the previous step (`h2` slots) and gives us the final answer (`out_features` slots). In our case, since we're trying to figure out which type of iris flower it is (Setosa, Versicolor, or Virginica), `out_features` is 3, one slot for each type. This is where the machine finally says, "Okay, this ball looks like it should go in the red bin!"

### Running the Machine (`forward`)

The `forward` part is like actually putting the balls into the sorting machine and watching them go through the steps.

*   `x = F.relu(self.fc1(x))`: We take the numbers (`x`) and send them through the first sorting step (`self.fc1`). Then, we use `F.relu`. Think of `F.relu` as a little helper that makes a decision: if a number is positive, it lets it through as is; if it's negative, it changes it to zero. This helps our machine learn more complicated ways to sort.

*   `x = F.relu(self.fc2(x))`: The numbers that came out of the first step now go through the second sorting step (`self.fc2`), and again, our `F.relu` helper makes decisions on those numbers.

*   `x = self.out(x)`: Finally, the numbers go through the last step, the output layer (`self.out`), which gives us our final result – the machine's best guess about what the input numbers represent.

*   `return x`: The machine gives us the final sorted result.

So, our neural network model is like a smart sorting machine with different steps (layers) that learn how to process numbers using decision-making helpers (`F.relu`) to give us a final answer.
```
